```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
df.head()
```

```python
# Made a copy for the analysis :
df_samp = df.copy()
df_samp.head(2)
```

```python
# Parsing Date and setting as index : 
df_samp['Date'] = pd.to_datetime(df_samp['Date'], format="%m/%d/%Y")
df_samp.set_index('Date', inplace=True)
```

```python
# Adding a ‚ÄúUnits_Sold‚Äù column (one sale per row)
df_samp['Units_Sold'] = 1
```

```python
# Quickly checking the data
print(df_samp.index.min(), df_samp.index.max())
```

```python
df_samp.head()
```

```python
# Daily + Weekly + Monthly Aggregation :
# Daily totals
daily = df_samp.resample('D').agg({
    'Units_Sold': 'sum',
    'Price ($)': 'sum'
}).rename(columns={'Price ($)': 'Daily_Revenue'})

# Weekly totals (Mon‚ÄìSun)
weekly = df_samp.resample('W').agg({
    'Units_Sold': 'sum',
    'Price ($)': 'sum'
}).rename(columns={'Price ($)': 'Weekly_Revenue'})

# Monthly totals
monthly = df_samp.resample('M').agg({
    'Units_Sold': 'sum',
    'Price ($)': 'sum'
}).rename(columns={'Price ($)': 'Monthly_Revenue'})

# Yearly totals
yearly = df_samp.resample('Y').agg({
    'Units_Sold': 'sum',
    'Price ($)': 'sum'
}).rename(columns={'Price ($)': 'Yearly_Revenue'})

# Inspect
print(daily.head(), weekly.head(), monthly.head(), yearly.head(), sep="\n\n")
```

```python
# Plotting the Aggregation :
plt.figure(figsize=(12,4))
plt.plot(daily['Units_Sold'], label='Daily Units Sold')
plt.plot(weekly['Units_Sold'], label='Weekly Units Sold', alpha=0.8)
plt.plot(monthly['Units_Sold'], label='Monthly Units Sold', linewidth=2)
plt.legend(); plt.title("Units Sold Over Time")
plt.show()

plt.figure(figsize=(12,4))
plt.plot(daily['Daily_Revenue'], label='Daily Revenue')
plt.plot(weekly['Weekly_Revenue'], label='Weekly Revenue', alpha=0.8)
plt.plot(monthly['Monthly_Revenue'], label='Monthly Revenue', linewidth=2)
plt.legend(); plt.title("Revenue Over Time")
plt.show()
```

```python
# Seasonality and Trend Decomposition :
# Decompose the monthly revenue into trend, seasonality, residual
decomp = seasonal_decompose(monthly['Monthly_Revenue'], model='additive')
decomp.plot()
plt.suptitle("Additive Decomposition of Monthly Revenue", y=1.02)
plt.show()
```

```python
# A basic monthly forecast (Using Holt-Winters)
# Fit Holt-Winters on monthly revenue
hw_model = ExponentialSmoothing(
    monthly['Monthly_Revenue'],
    trend='add',
    seasonal='add',
    seasonal_periods=12
).fit()

# Forecast next 6 months
forecast = hw_model.forecast(4)

# Combine for plotting
proj = pd.concat([monthly['Monthly_Revenue'], forecast], axis=0)

plt.figure(figsize=(10,4))
plt.plot(monthly['Monthly_Revenue'], label='Historical')
plt.plot(forecast, label='Forecast', linestyle='--', marker='o')
plt.title("3-Month Revenue Forecast (Holt‚ÄìWinters)")
plt.legend()
plt.show()
```

```python
# Top Brands & Models by Sales Volume and Revenue
# 1. Aggregate by Company
brand_summary = (
    df_samp
    .groupby('Company')
    .agg(
        Units_Sold=('Units_Sold', 'sum'),
        Total_Revenue=('Price ($)', 'sum')
    )
    .sort_values(by='Units_Sold', ascending=False)
)

print("Top Brands by Units Sold & Revenue:")
print(brand_summary.head(10))

# 2. Aggregate by Company + Model
model_summary = (
    df_samp
    .groupby(['Company', 'Model'])
    .agg(
        Units_Sold=('Units_Sold', 'sum'),
        Total_Revenue=('Price ($)', 'sum')
    )
    .sort_values(by='Units_Sold', ascending=False)
)

print("\nTop Models by Units Sold & Revenue:")
print(model_summary.head(10))
```
